```python
import pandas as pd
import json
import requests

url_hospedagem = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_hospedagem.json'

response = requests.get(url_hospedagem)

dados = pd.DataFrame(response.json())
dados.head()
```

```python
#Normalização: json_normalize para "normalizar" um objeto, buscando converter os dados dele para um formato mais estruturado, como um DataFrame.
dados = pd.json_normalize(dados['info_moveis'])
dados
```

# Manipulação de dados textuais para criação de uma estrutura de tokenização simples
## Tokenização = Dividir o texto em unidades menores


```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

```python
# Removendo caracteres especiais com regex
# ^ ->  Sinal de negação. Significa que vai selecionar todos os caracteres menos os que serão digitados em seguida
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True)
```

```python
# primeira e a segunda parte do regex verifica se não há algum caractere antes e depois do hífen
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex = True)
```

```python
dados['descricao_local'].str.split()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades'].str.replace('\{|}|\"', '', regex = True)
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex = True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```
